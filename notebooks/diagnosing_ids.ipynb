```python
all_res = sLD1.aggregate([
    {"$match": {'Type': regx}}, 
    {"$group": 
     { "_id": 
      { "RegistrantID": "$Registrant.RegistrantID", 
        "ClientID": "$Client.ClientID", 
        "Year": "$Year"}, 
      "count": {"$sum": 1}}},
    {"$project": 
     { "RegistrantID": "$_id.RegistrantID", 
       "ClientID": "$_id.ClientID", 
       "Year": "$_id.Year",
       "count": "$count", 
       "_id": 0}}])
```

```python
sopr_sen_all = pd.DataFrame(all_res['result'])
```
