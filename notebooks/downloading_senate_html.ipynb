```python
def get_left_bounds(factor):
    return [int(eval(s.split(',')[0].replace('[','').replace('(',''))) for s in factor.levels.values]
```

```python
bounds = []
counts = []

for day in big_days:
    group = gb.get_group(day)
    factor = pd.qcut(group[group>0], 5)
    counts.append(factor.describe()['counts'].values)
    bounds.append(get_left_bounds(factor))
```

```python
def simulate_cuts(cuts, days, grouped):
    counts = []
    colnames = ['B-'+str(c) for c in cuts[:-1]]
    for day in days:
        group = grouped.get_group(day)
        out, bins = pd.cut(group[group>0], cuts, retbins=True, labels=False)
        print bins
        counts.append(np.bincount(out))
    return pd.DataFrame(counts, index=big_days, 
                        columns=colnames)
```
